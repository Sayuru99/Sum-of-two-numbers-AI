```python
import random
import numpy as np
import pandas as pd
import math
```

**Creating dataset**  
Creating random numbers in 1 to 1000 range

```python
X=[]
Y=[]

for i in range(1000):
  X.append([random.randint(1,1000), random.randint(1,1000)])
  Y.append(sum(X[i]))

X = np.array(X)
Y = np.array(Y).reshape(-1,1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

Creating layers

```python
model.add(Dense(6, activation='relu', input_dim=2))
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='linear')) #used linear activation because this is a linear regression problem
```

```python
model.compile('adam', 'mean_squared_error')
model.fit(X, Y, epochs=200)
```

**Testing**  
Input two numbers to get sum

```python
pred = np.array([[80,20]])
pred2 = model.predict(pred)
add = math.ceil(pred2) #rounds number up
print(add)
```
